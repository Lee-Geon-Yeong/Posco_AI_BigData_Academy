### 실습 2)

```python
# mycars.csv 데이터 가져오기
ds_mycars=pd.read_csv("C:\\Users\\Administrator\\Untitled Folder\\statistics_1day/mycars.csv", engine="python")
# df에 ds_mycars의 model, mpg의 column 데이터를 저장합니다.
df=ds_mycars[['model', 'mpg']]

# model 변수 Count: value_count, 순서대로 정렬: sort_index
count=df['model'].value_counts().sort_index()

# CumCnt 계산: cumsum
percent=count/sum(count)*100
# Percent 계산: 직접
cumcnt=np.cumsum(count)
# CumPct 계산: cumsum
cumpct=np.cumsum(percent)

# DataFrame으로 취합
count_data=pd.DataFrame({'Count':count, 'CumCnt':cumcnt, 'Percent':percent, 'CumPct:':cumpct})

# Column의 name(좌측상단 이름) 생성
count_data.columns.name='model'
count_data
```

```python
# mpg 변수 Count: value_count, 순서대로 정렬: sort_index
count2=df['mpg'].value_counts().sort_index()
# CumCnt 계산: cumsum
cumcnt2=np.cumsum(count2)
# percent 계산: 직접
percent2=count2/sum(count2)*100
# CumPct 계산: cumsum
cumpct2=np.cumsum(percent2)
# DataFrame으로 취합
count_data2=pd.DataFrame({'Count':count2, 'CumCnt':cumcnt2, 'Percent':percent2, 'CumPct:':cumpct2})
# Column의 name(좌측상단 이름) 생성
count_data2.columns.name='mpg'
count_data2
```

#### 고찰)
#### 이전 실습 1)과의 차이점은 DataFrame 형식으로 2차원 배열에 저장을 한다는 점입니다.
#### pandas에는 Series(1차원) - index, value 와 DataFrame(2차원) - index, column, value 의 두 가지 데이터 형식이 있습니다.
#### pandas를 쓰는 이유는 전처리 관련한 내장함수가 많고, Series와 DataFrame이라는 데이터 형식이 있기 때문에 널리 쓰입니다.
#### DataFrame을 딕셔너리 형태로 불러왔는데, 딕셔너리 형태를 쓰는 이유는 가중치를 저장하기 쉽고, pandas로 엑셀 변환을 쉽게 하기 위해 쓰입니다.

### ※ 기술통계 표를 왜 구하고 분석하는가?
#### 기술통계 표를 이용해 평균(mean)과 표준편차(std)를 구하고, 변동계수(cv)=표준편차/평균 을 구할 수 있습니다.
#### 이상치의 값을 갖는 X의 값이 목표변수 Y의 값에 영향을 줄 가능성이 크다고 판단하여, 표준화하여 오차를 줄일 수 있습니다.
